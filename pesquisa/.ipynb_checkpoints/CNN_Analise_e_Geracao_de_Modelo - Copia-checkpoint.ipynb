## Classificando ovinos doentes e saudáveis em rebanhos

O objetivo deste projeto é classificar corretamente ovinos como doentes ou saudáveis por meio de imagens RGB da mucosa do olho dos animais.

Os animais foram pré-classificados de acordo com o resultado do exame de contagem de hematócritos, seguindo o padrão abaixo:

* Abaixo de 24: Doente
* Igual ou acima de 24: Saudável

### Por que é importante? 

De acordo com a justificativa do projeto:

"A alta taxa de incidência de anemia em caprinos e ovinos gera uma alta demanda de testagem nos animais, podendo esta frequência chegar a ser semanal. Desta forma, de acordo com o tamanho do rebanho, pode ser um trabalho repetitivo e altamente dispendioso de tempo e recursos para o produtor, além do fato de aumentar consideravelmente as chances de erro humano, aumentando ainda mais os prejuízos dos produtores."


### Qual é o objetivo do modelo de Machine Learning?

Maximizar a acurácia, isto é, a classificação correta dos diferentes tipos de ovinos (saudáveis e doentes).

## Carrega as Imagens / Dados

```python
# Aqui estão todas as nossas imagens
DATA_FOLDER = 'imagens'
```

```python
# Cada classe de animal está em uma pasta dedicada
os.listdir(DATA_FOLDER)
```

```python
# Vamos percorrer toda a estrutura de diretórios da pasta, de modo que possamos entender
# como as imagens estão armazenadas

for class_folder_name in os.listdir(DATA_FOLDER):
    class_folder_path = os.path.join(DATA_FOLDER, class_folder_name)
    for image_path in glob(os.path.join(class_folder_path, "*.jpg")):
        print(image_path)
```

```python
# Criação do Dataframe:
# o dataframe armazena todos os caminhos de acesso às imagens em uma coluna
# e a classe do animal (o alvo) na coluna seguinte

images_df = []

# Navega dentro de cada pasta
for class_folder_name in os.listdir(DATA_FOLDER):
    class_folder_path = os.path.join(DATA_FOLDER, class_folder_name)
    
    # Coleta o caminho de cada imagem
    for image_path in glob(os.path.join(class_folder_path, "*.jpg")):
        
        tmp = pd.DataFrame([image_path, class_folder_name]).T
        images_df.append(tmp)

# Concatena o dataframe final
images_df = pd.concat(images_df, axis=0, ignore_index=True)
images_df.columns = ['image', 'target']
images_df.head(10)
```

```python
# Aqui é feita a plotagem de um conjunto de imagens juntas,
# de modo a ter uma visão melhor de como são as imagens de cada classe

def plot_for_class(df, label):
    # esta função plota as 9 primeiras imagens de cada classe
    nb_rows = 3
    nb_cols = 3
    
    fig, axs = plt.subplots(nb_rows, nb_cols, figsize=(10, 10))

    n = 0
    for i in range(0, nb_rows):
        for j in range(0, nb_cols):
            tmp = df[df['target'] == label ]
            tmp.reset_index(drop=True, inplace=True)
            im = cv2.imread(tmp.loc[n, 'image'])          
            axs[i, j].imshow(im)
            n += 1 
```

## Separando os conjuntos de dados de treino e teste

```python
# Separa o conjunto de dados em treino e teste (train_test_split)

X_train, X_test, y_train, y_test = train_test_split(images_df['image'], images_df['target'], test_size=0.20, random_state=101)

print(X_train.shape)
print(X_test.shape)
```

```python
# os índices dos dados de treino estão fora de ordem
# isso irá certamente causar problemas futuros, pois precisaremos iterar usando o número da linha.

X_train.head()
```

```python
# Para resolver esse problema, atribuiremos novos índices aos dados de treino e teste

X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
X_train.head()
```

```python
# Atribuiremos novos índices no alvo também

y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
# Porcentagem de imagens dentro de cada classe

y_train.value_counts() / len(y_train)
```

```python
# Deve ser similar no conjunto de teste

y_test.value_counts() / len(y_test)
```

```python
# Vamos preparar o alvo
# Esta é uma classificação de múltiplas classes, deste modo vamos aplicar o método 
# one-hot-encoding para transformar a variável alvo em colunas distintas

encoder = LabelEncoder()
encoder.fit(y_train)

train_y = np_utils.to_categorical(encoder.transform(y_train))
test_y = np_utils.to_categorical(encoder.transform(y_test))

train_y
```

```python
# As imagens em nossas pastas podem ser de diferentes tamanhos
# Porém, para trabalharmos com Redes Neurais, faz-se necessários que as imagens tenham o mesmo tamanho
# Dessa forma, as imagens serão todas redimensionadas para o tamanho abaixo:

IMAGE_SIZE = 150
```

```python
def im_resize(df, n):
    im = cv2.imread(df[n])
    im = cv2.resize(im, (IMAGE_SIZE, IMAGE_SIZE))
    return im
```

```python
tmp = im_resize(X_train, 7)
tmp.shape
```

```python
# O formato dos conjuntos de dados deve ser (n1, n2, n3, n4)
# onde n1 é o número de observações
# n2 e n3 são a largura e o comprimento das imagens
# e n4 indica o número de planos por imagem; 
# como estas são imagens coloridas (RGB), então precisaremos ter 3 planos por imagem

def create_dataset(df, image_size):
    # cria o conjunto de dados de acordo com o exigido para uma cnn
    tmp = np.zeros((len(df), image_size, image_size, 3), dtype='float32')

    for n in range(0, len(df)):
        im = im_resize(df, n)
        tmp[n] = im
  
    print('Formato das imagens do conjunto de dados: {} tamanho: {:,}'.format(tmp.shape, tmp.size))
    return tmp
```

```python
x_train = create_dataset(X_train, IMAGE_SIZE)
```

```python
x_test = create_dataset(X_test, IMAGE_SIZE)
```

```python
# Número de classes diferentes

len(y_train.unique())
```

```python
# Essa é nossa Rede Convolucional

kernel_size = (3,3)
pool_size= (2,2)
first_filters = 32
second_filters = 64
third_filters = 128

dropout_conv = 0.3
dropout_dense = 0.3

model = Sequential()
model.add(Conv2D(first_filters, kernel_size, activation = 'relu', 
                 input_shape = (IMAGE_SIZE, IMAGE_SIZE, 3)))
model.add(Conv2D(first_filters, kernel_size, activation = 'relu'))
#model.add(Conv2D(first_filters, kernel_size, activation = 'relu'))
model.add(MaxPooling2D(pool_size = pool_size)) 
model.add(Dropout(dropout_conv))

model.add(Conv2D(second_filters, kernel_size, activation ='relu'))
model.add(Conv2D(second_filters, kernel_size, activation ='relu'))
#model.add(Conv2D(second_filters, kernel_size, activation ='relu'))
model.add(MaxPooling2D(pool_size = pool_size))
model.add(Dropout(dropout_conv))

model.add(Conv2D(third_filters, kernel_size, activation ='relu'))
model.add(Conv2D(third_filters, kernel_size, activation ='relu'))
#model.add(Conv2D(third_filters, kernel_size, activation ='relu'))
model.add(MaxPooling2D(pool_size = pool_size))
model.add(Dropout(dropout_conv))

model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(dropout_dense))
model.add(Dense(2, activation = "softmax"))

model.summary()
```

```python
model.compile(Adam(lr=0.0001), loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
batch_size = 10
epochs = 8
```

```python
x_train.shape
```

```python
filepath = "model.h5"
checkpoint = ModelCheckpoint(filepath, monitor='accuracy', verbose=1, 
                             save_best_only=True, mode='max')

reduce_lr = ReduceLROnPlateau(monitor='accuracy', factor=0.5, patience=1, 
                                   verbose=1, mode='max', min_lr=0.00001)
                              
                              
callbacks_list = [checkpoint, reduce_lr]

history = model.fit(x=x_train, y=train_y,
                    batch_size=batch_size, 
                    epochs=epochs,
                    
                    verbose=2,
                    callbacks=callbacks_list)
```

```python
acc = history.history['accuracy']
loss = history.history['loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, loss, 'bo', label='Perda ao longo do treinamento')
plt.title('Perda ao longo do treinamento')
plt.legend()
plt.figure()

plt.plot(epochs, acc, 'bo', label='Precisão ao longo do treinamento')
plt.title('Precisão ao longo do treinamento')
plt.legend()
plt.figure()
```

```python
# Faz uma previsão

predictions = np.argmax(model.predict(x_test),axis=1)
```

```python
# Verifica a previsão
predictions
```

```python
# Transforma o alvo
encoder.transform(y_test)
```

```python
cnf_matrix = confusion_matrix(encoder.transform(y_test), predictions)
```

```python
# Cria um dicionário para mapear de volta os números em nomes de classes legíveis

tmp = pd.concat([y_test, pd.Series(encoder.transform(y_test))], axis=1)
tmp.columns = ['class_name', 'class_number']
tmp.drop_duplicates(inplace=True)
tmp.reset_index(drop=True, inplace=True)
tmp = pd.Series(tmp.class_name.values,index=tmp.class_number).to_dict()
tmp
```
